### time DF 가공하기

- time 수정하기 : 보기편하도록 시간을 숫자로 보기 변하도록 변경함

```python
time_colname_list = ['MIDNIGHT_RIDE_NUM', 'MIDNIGHT_ALIGHT_NUM','ONE_RIDE_NUM', 'ONE_ALIGHT_NUM', 'TWO_RIDE_NUM', 'TWO_ALIGHT_NUM',
       'THREE_RIDE_NUM', 'THREE_ALIGHT_NUM', 'FOUR_RIDE_NUM','FOUR_ALIGHT_NUM', 'FIVE_RIDE_NUM', 'FIVE_ALIGHT_NUM', 'SIX_RIDE_NUM',
       'SIX_ALIGHT_NUM', 'SEVEN_RIDE_NUM', 'SEVEN_ALIGHT_NUM','EIGHT_RIDE_NUM', 'EIGHT_ALIGHT_NUM', 'NINE_RIDE_NUM',
       'NINE_ALIGHT_NUM', 'TEN_RIDE_NUM', 'TEN_ALIGHT_NUM', 'ELEVEN_RIDE_NUM','ELEVEN_ALIGHT_NUM', 'TWELVE_RIDE_NUM', 'TWELVE_ALIGHT_NUM',
       'THIRTEEN_RIDE_NUM', 'THIRTEEN_ALIGHT_NUM', 'FOURTEEN_RIDE_NUM','FOURTEEN_ALIGHT_NUM', 'FIFTEEN_RIDE_NUM', 'FIFTEEN_ALIGHT_NUM',
       'SIXTEEN_RIDE_NUM', 'SIXTEEN_ALIGHT_NUM', 'SEVENTEEN_RIDE_NUM','SEVENTEEN_ALIGHT_NUM', 'EIGHTEEN_RIDE_NUM', 'EIGHTEEN_ALIGHT_NUM',
       'NINETEEN_RIDE_NUM', 'NINETEEN_ALIGHT_NUM', 'TWENTY_RIDE_NUM','TWENTY_ALIGHT_NUM', 'TWENTY_ONE_RIDE_NUM', 'TWENTY_ONE_ALIGHT_NUM',
       'TWENTY_TWO_RIDE_NUM', 'TWENTY_TWO_ALIGHT_NUM', 'TWENTY_THREE_RIDE_NUM','TWENTY_THREE_ALIGHT_NUM']

corrected_colname_list = []
ride_colname_list = []
alight_colname_list = []
for i in range(24):
    corrected_colname_list.append('ride_' + str(i))
    corrected_colname_list.append('alight_' + str(i))
    ride_colname_list.append('ride_' + str(i))
    alight_colname_list.append('alight_' + str(i))

time.rename(columns = dict(zip(time_colname_list, corrected_colname_list)), inplace = True)
time.head(1)
```

```python
time_df = time.copy()
```

- wide_to_long 함수 : 데이터 프레임을 "wide" 형식에서 "long" 형식으로 변환하는 데 사용

```python
time_df["idx"] = time_df.index
time_df = pd.wide_to_long(time_df, stubnames=['ride', 'alight'], i=['idx'],j='TIME', sep='_')
time_df
```

```python
time_df1 = time_df.reset_index()
time_df1.drop(['idx', 'WORK_DT'], axis=1, inplace = True)
time_df1.rename(columns={'ride': 'RIDE_NUM', 'alight': 'ALIGHT_NUM'}, inplace = True)
time_df1
```

- 사용년월,노선번호,표준버스정류장ID가 중복되는 경우

```python
temp_grouped = time_df1.groupby(['STND_BSST_ID', 'BUS_ROUTE_NO', 'USE_MON']).size()
temp_grouped.unique() #고유값
```

```python
time_df1[ (time_df1['STND_BSST_ID'] == 100000001) &  (time_df1['BUS_ROUTE_NO'] == 'N37')&  (time_df1['USE_MON'] == 202301)]
```

```python
#  cols에 해당하는 값들이 다 같으면 중복으로 간주하고 첫번째 행만 유지
cols = ['TIME', 'BUS_STA_NM', 'BUS_ROUTE_NM', 'STND_BSST_ID', 'BSST_ARS_NO', 'BUS_ROUTE_NO', 'USE_MON']
time_df1 = time_df1.drop_duplicates( cols, keep="first" )
time_df1
```

- 총 승차 인원 수를 추가

```python
ride_sum_grouped = time_df1.groupby(['STND_BSST_ID', 'BUS_ROUTE_NO', 'USE_MON']).agg({'RIDE_NUM': 'sum'}) #그룹화하여, ride_num열값을 합산
time_df2 = time_df1.merge(ride_sum_grouped, how='left', on=['STND_BSST_ID', 'BUS_ROUTE_NO', 'USE_MON']) #병합
```

```python
time_df2[ (time_df1['STND_BSST_ID'] == 100000001) &  (time_df1['BUS_ROUTE_NO'] == 'N37')&  (time_df2['USE_MON'] == 202301)]
```

```python
#승차 인원이 그룹 내 전체 승차 인원에서 차지하는 비율을 계산
time_df2['RIDE_PCT'] = round(time_df2['RIDE_NUM_x'] / time_df2['RIDE_NUM_y'], 4)
time_df2
```

```python
#하차 인원이 그룹 내 전체 인원에서 차지하는 비율을 계산
alight_sum_grouped = time_df1.groupby(['STND_BSST_ID', 'BUS_ROUTE_NO', 'USE_MON']).agg({'ALIGHT_NUM': 'sum'})
time_df2 = time_df2.merge(alight_sum_grouped, how='left', on=['STND_BSST_ID', 'BUS_ROUTE_NO', 'USE_MON'])
time_df2['ALIGHT_PCT'] = round(time_df2['ALIGHT_NUM_x'] / time_df2['ALIGHT_NUM_y'], 4)
time_df2
```

```python
#하루동안 0이라 NAN값 나오는 것들 처리
print(time_df2.isnull().sum())
time_df2['RIDE_PCT'] = time_df2['RIDE_PCT'].fillna(0)
time_df2['ALIGHT_PCT'] = time_df2['ALIGHT_PCT'].fillna(0)
print(time_df2.isnull().sum())
```

```python
# col 이름 바꿔주기
col_dict = {'RIDE_NUM_x': 'RIDE_NUM', 'ALIGHT_NUM_x': 'ALIGHT_NUM', 'RIDE_NUM_y': 'RIDE_SUM', 'ALIGHT_NUM_y': 'ALIGHT_SUM'}
time_df2.rename(columns=col_dict, inplace=True)
time_df2
```

```python
# col 순서 정리
cols = ['BUS_ROUTE_NO', 'BUS_ROUTE_NM', 
        'STND_BSST_ID', 'BSST_ARS_NO', 'BUS_STA_NM',
        'USE_MON', 'TIME',
        'RIDE_NUM', 'RIDE_PCT',	'ALIGHT_NUM', 'ALIGHT_PCT', 'RIDE_SUM', 'ALIGHT_SUM']  #  'RIDE_SUM', 'ALIGHT_SUM'
time_df2 = time_df2[cols]
time_df2
```

```python
time_df2.to_csv('preproccessed_bus_by_time.csv', index=False)
```

## date data 가공하기

```python
date_df = date.copy()
date_df.drop(['WORK_DT'], axis=1, inplace=True)
date_df.head()
```

```python
date_df['USE_DT'] = pd.to_datetime(date_df['USE_DT'], format='%Y%m%d', errors='raise')
date_df['WEEKDAY'] = date_df['USE_DT'].dt.weekday
date_df
```

```python
# col 이름 바꿔주기
col_dict = {'RIDE_PASGR_NUM': 'RIDE_NUM', 'ALIGHT_PASGR_NUM': 'ALIGHT_NUM'}
date_df.rename(columns=col_dict, inplace=True)
```

```python
# col 순서 정리
cols = ['BUS_ROUTE_NO', 'BUS_ROUTE_NM', 'BUS_ROUTE_ID',
        'STND_BSST_ID', 'BSST_ARS_NO', 'BUS_STA_NM',
        'USE_DT', 'WEEKDAY',
        'RIDE_NUM', 'ALIGHT_NUM']
date_df = date_df[cols]
date_df
```

```python
date_df.to_csv('preprocessed_bus_by_date.csv', index=False)
```
